In what ways have the Olympics Games become more progressive with gender overtime?

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/athlete_events.csv.zip')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.drop_duplicates(inplace = True)
```

```python
data.isna()
```

```python
data.drop(columns = ['Height', 'Weight'], inplace = True)
```

```python
data['Medal'] = data['Medal'].fillna('None')
data.head()
```

```python
data.head()
```

```python
from matplotlib import pyplot as plt
data.plot(kind='scatter', x='Sex', y='Year', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
plt.title('Gender vs. Year')
```

```python
# @title Sex

from matplotlib import pyplot as plt
import seaborn as sns
data.groupby('Sex').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
medals = data.dropna(subset = ['Medal'])
medals_by_country = medals['NOC'].value_counts()
top_countries = medals_by_country.head(10)
top_countries.plot(kind='bar')

plt.title('Medals Won for All Countries')
plt.xlabel('Team')
plt.ylabel('Medal')
plt.xticks(rotation = 90)
plt.show()
```

```python
from operator import ge
data_unique = data.drop_duplicates(subset = ['Name','NOC','Games','Sex'])
gender_counts = data_unique.groupby(['Team', 'Sex']).size().unstack(fill_value=0)
gender_counts['Total']= gender_counts['F'] + gender_counts['M']
gender_counts = gender_counts.sort_values(by='Total', ascending=False).drop(columns='Total')
gender_counts.head(10).plot(kind='bar')
plt.show()
```
